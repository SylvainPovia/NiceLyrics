```python

import stanfordnlp
```

```python
# Import le fichier XML
import xml.etree.ElementTree as ET
tree = ET.parse('Lyric_song.xml')
root = tree.getroot()
xmlstr = ET.tostring(root, encoding='utf8', method='xml')
# all items data
print('Expertise Data:')
text_test = ""
for elem in root:
    for subelem in elem:
        text_test = text_test + subelem.text
        print(subelem.text)
```

```python
# Telecharge le modele StanfordNLP en Français
# Peut prendre un peu de temps
#stanfordnlp.download('fr')
```

```python
nlp = stanfordnlp.Pipeline(lang="fr",processors = "tokenize,mwt,lemma,pos")
```

```python
test = text_test.replace("\n", " ")
```

```python
doc = nlp(test)
for i in range(len(doc.sentences)):
    doc.sentences[i].print_tokens()
```
